# **Pauli Matrix Trace Properties**

This notebook demonstrates and proves key properties of the trace operation for Pauli matrices and general matrices.

Case 1, $i=j$, $2\delta_{ij} = 2$

\begin{align*}
    Tr(\begin{bmatrix}
        0 & 1 \\
        1 & 0
    \end{bmatrix}\begin{bmatrix}
        0 & 1 \\
        1 & 0
    \end{bmatrix}) &= Tr(\begin{bmatrix}
        1 & 0 \\
        0 & 1
    \end{bmatrix}) = 2 \\
    Tr(\begin{bmatrix}
        0 & -i \\
        i & 0
    \end{bmatrix}\begin{bmatrix}
        0 & -i \\
        i & 0
    \end{bmatrix}) &= Tr(\begin{bmatrix}
        1 & 0 \\
        0 & 1
    \end{bmatrix}) = 2 \\
    Tr(\begin{bmatrix}
        1 & 0 \\
        0 & -1
    \end{bmatrix}\begin{bmatrix}
        1 & 0 \\
        0 & -1
    \end{bmatrix}) &= Tr(\begin{bmatrix}
        1 & 0 \\
        0 & 1
    \end{bmatrix}) = 2 
\end{align*}

**Properties of the Trace**

*1. $Tr(AB) = Tr(BA)$*

Let $A$ and $B$ be $N\times N$ matricies such that
\begin{equation*}
    A = \begin{bmatrix}
        a_{11} & a_{12} & \cdots & a_{1N} \\
        a_{21} & a_{22} & \cdots & a_{2N} \\
        \vdots & \vdots & \ddots & \vdots \\
        a_{N1} & a_{N2} & \cdots & a_{NN}
    \end{bmatrix}, \quad
    B = \begin{bmatrix}
        b_{11} & b_{12} & \cdots & b_{1N} \\
        b_{21} & b_{22} & \cdots & b_{2N} \\
        \vdots & \vdots & \ddots & \vdots \\
        b_{N1} & b_{N2} & \cdots & b_{NN}
    \end{bmatrix}.
\end{equation*}

The product $AB$ is
\begin{equation*}
    AB = \begin{bmatrix}
        \sum_{i=1}{N} a_{1i}b_{i1} & \sum_{i=1}^{N} a_{1i}b_{i2} & \cdots & \sum_{i=1}^{N} a_{1i}b_{iN} \\
        \sum_{i=1}^{N} a_{2i}b_{i1} & \sum_{i=1}^{N} a_{2i}b_{i2} & \cdots & \sum_{i=1}^{N} a_{2i}b_{iN} \\
        \vdots & \vdots & \ddots & \vdots \\
        \sum_{i=1}^{N} a_{Ni}b_{i1} & \sum_{i=1}^{N} a_{Ni}b_{i2} & \cdots & \sum_{i=1}^{N} a_{Ni}b_{iN}
    \end{bmatrix}.
\end{equation*}

The trace of $AB$ is 
\begin{equation*}
    Tr(AB) = \sum_{i=1}^{N} \left( \sum_{j=1}^{N} a_{ij}b_{ji} \right) = \sum_{i=1}^{N} \sum_{j=1}^{N} a_{ij}b_{ji}.
\end{equation*}

Similarly, the product $BA$ is 
\begin{equation*}
    BA = \begin{bmatrix}
        \sum_{i=1}^{N} b_{1i}a_{i1} & \sum_{i=1}^{N} b_{1i}a_{i2} & \cdots & \sum_{i=1}^{N} b_{1i}a_{iN} \\
        \sum_{i=1}^{N} b_{2i}a_{i1} & \sum_{i=1}^{N} b_{2i}a_{i2} & \cdots & \sum_{i=1}^{N} b_{2i}a_{iN} \\
        \vdots & \vdots & \ddots & \vdots \\
        \sum_{i=1}^{N} b_{Ni}a_{i1} & \sum{j=1}^{N} b_{Ni}a_{ij2} & \cdots & \sum_{j=1}^{N} b_{Ni}a_{ijN}
    \end{bmatrix}.
\end{equation*}

The trace of $BA$ is 
\begin{equation*}
    Tr(BA) = \sum_{i=1}^{N} \left( \sum_{j=1}^{N} b_{ij}a_{ji} \right) = \sum_{i=1}^{N} \sum_{j=1}^{N} b_{ij}a_{ji}
\end{equation*}
since the indices $i$ and $j$ are dummy variables, $Tr(AB) = Tr(BA)$.

*2. $Tr(ABC) = Tr(CAB) = Tr(BCA)$*

Let $A$, $B$, and $C$ be $N\times N$ matricies. From part 1, we have the product $AB$. The product $ABC$ is given by
\begin{equation*}
    ABC = \begin{bmatrix}
        \sum_{i=1}^N \sum_{j=1}^N a_{1i}b_{ij}c_{j1} & \sum_{i=1}^N \sum_{j=1}^N a_{1i}b_{ij}c_{j2} & \cdots & \sum_{i=1}^N \sum_{j=1}^N a_{1i}b_{ij}c_{jN} \\
        \sum_{i=1}^N \sum_{j=1}^N a_{2i}b_{ij}c_{j1} & \sum_{i=1}^N \sum_{j=1}^N a_{2i}b_{ij}c_{j2} & \cdots & \sum_{i=1}^N \sum_{j=1}^N a_{2i}b_{ij}c_{jN} \\
        \vdots & \vdots & \ddots & \vdots \\
        \sum_{i=1}^N \sum_{j=1}^N a_{Ni}b_{ij}c_{j1} & \sum_{i=1}^N \sum_{j=1}^N a_{Ni}b_{ij}c_{j2} & \cdots & \sum_{i=1}^N \sum_{j=1}^N a_{Ni}b_{ij}c_{jN}
    \end{bmatrix}.
\end{equation*}

The trace of $ABC$ is thus
\begin{equation*}
    Tr(ABC) = \sum_{k=1}^{N} \left( \sum_{i=1}^{N} \sum_{j=1}^{N} a_{ki}b_{ij}c_{jk} \right).
\end{equation*}

The traces of $CAB$ and $BCA$ can be calculated similarly:
\begin{align*}
    Tr(CAB) &= \sum_{k=1}^{N} \left( \sum_{i=1}^{N} \sum_{j=1}^{N} c_{ki}a_{ij}b_{jk} \right), \\
    Tr(BCA) &= \sum_{k=1}^{N} \left( \sum_{i=1}^{N} \sum_{j=1}^{N} b_{ki}c_{ij}a_{jk} \right).
\end{align*}

Again, since $i$, $j$, and $k$ are dummy variables, $Tr(ABC) = Tr(CAB) = Tr(BCA)$.

*3. $Tr(A\otimes b) = Tr(A)Tr(B)$*

Let $A$ and $B$ both be $N\times N$ matricies as defined previously. The tensor product of the two matricies has diagonal elements:
\begin{equation*}
    A\otimes B = \begin{bmatrix}
        a_{11}B & \quad & \quad & \quad \\
        \quad & a_{22}B & \quad & \quad \\
        \quad & \quad & \ddots & \quad \\
        \quad & \quad & \quad & a_{NN}B
    \end{bmatrix}.
\end{equation*}

The trace of this tensor product is just the sum of the traces of the diagonal elements:
\begin{align*}
    Tr(A\otimes B) &= a_{11}b_{11} + a_{11}b_{22} + \cdots + a_{22}b_{11} + a_{22}b_{22} + \cdots \\
    &= a_{11}Tr(B) + a_{22}Tr(B) + \cdots \\
    &= Tr(A)Tr(B).
\end{align*}
