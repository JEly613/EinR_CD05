Case 1, $i=j$, $2\delta_{ij} = 2$

\begin{align*}
    Tr(\begin{bmatrix}
        0 & 1 \\
        1 & 0
    \end{bmatrix}\begin{bmatrix}
        0 & 1 \\
        1 & 0
    \end{bmatrix}) &= Tr(\begin{bmatrix}
        1 & 0 \\
        0 & 1
    \end{bmatrix}) = 2 \\
    Tr(\begin{bmatrix}
        0 & -i \\
        i & 0
    \end{bmatrix}\begin{bmatrix}
        0 & -i \\
        i & 0
    \end{bmatrix}) &= Tr(\begin{bmatrix}
        1 & 0 \\
        0 & 1
    \end{bmatrix}) = 2 \\
    Tr(\begin{bmatrix}
        1 & 0 \\
        0 & -1
    \end{bmatrix}\begin{bmatrix}
        1 & 0 \\
        0 & -1
    \end{bmatrix}) &= Tr(\begin{bmatrix}
        1 & 0 \\
        0 & 1
    \end{bmatrix}) = 2 
\end{align*}

Case 2, $i \neq j$, $2\delta_{ij} = 0$:

\begin{align*}
    Tr(\begin{bmatrix}
        0 & 1 \\
        1 & 0
    \end{bmatrix}\begin{bmatrix}
        0 & -i \\
        i & 0
    \end{bmatrix}) &= Tr(\begin{bmatrix}
        -i & 0 \\
        0 & i
    \end{bmatrix}) = 0 \\
    Tr(\begin{bmatrix}
        0 & 1 \\
        1 & 0
    \end{bmatrix}\begin{bmatrix}
        1 & 0 \\
        0 & -1
    \end{bmatrix}) &= Tr(\begin{bmatrix}
        0 & -1 \\
        -1 & 0
    \end{bmatrix}) = 0 \\
    Tr(\begin{bmatrix}
        0 & -i \\
        i & 0
    \end{bmatrix}\begin{bmatrix}
        1 & 0 \\
        0 & -1
    \end{bmatrix}) &= Tr(\begin{bmatrix}
        i & 0 \\
        0 & -i
    \end{bmatrix}) = 0
\end{align*}
